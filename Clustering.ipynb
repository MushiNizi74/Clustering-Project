# Task 2.2
Making a Kmeans Model

```python
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris #loading directly from sklearn datasets
iris= load_iris()
```

```python
kmeans= KMeans(n_clusters=3) #initializing number of clusters
```

```python
KModel = kmeans.fit(iris.data) #fitting the data
```

```python
centroids = KModel.cluster_centers_ #making centroids
centroids
```

```python
Kfit= kmeans.predict(iris.data) #prediction for data
Kfit 
```

```python
plt.figure()
plt.scatter(Kfit[0:],Kfit[:150])
```

# Task 2.3
Plotting the Iris Data Set

```python
from sklearn.mixture import GaussianMixture 
gmm = GaussianMixture(n_components = 3) #initializing centers
gmm.fit(data) #fitting data
labels = gmm.predict(data) #prediction data
cluster=gmm.means_ #centroid
colormap = np.array(['b', 'r', 'g']) #plotting of figure
plt.scatter(data[:,0],data[:,1], c = colormap[labels])
plt.scatter(cluster[:, 0], cluster[:, 1], c = 'black', marker = 'X', s=200)
plt.xlabel('Sepal Length(cm)')   
plt.ylabel('Sepal Width(cm)')
plt.title('The Iris Data Set')
plt.show() 
```

```python
prob=gmm.predict_proba(data) #shaping your data
print(prob.shape)
```

```python
for i in range(gmm.n_components): #putting loops for the three maps
  plt.figure()
  plt.scatter(data[:,0], data[:,1], c=prob[:,i]) 

  plt.ylabel('Sepal Width')
  plt.title('GMM differences', size=20)
  plt.colorbar() #putting colorbar
  plt.show()
```
